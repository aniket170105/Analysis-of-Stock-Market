# Preparing DataSet

```python
import pandas as pd
import yfinance as yf

stock_symbol = 'ACN'
timeframe = '5m'
test_stock = yf.download(stock_symbol, interval=timeframe, period = "1mo")
test_stock.drop(columns = ['Adj Close'], inplace = True)

stock = pd.read_csv(f'D:/Design Credit 1/Data Sector Wise/Technology/{stock_symbol}/{stock_symbol}TimeSeries-Final.csv')
stock.drop(columns = ['Unnamed: 0'], inplace = True)
test_stock
```

**Calculating Indicators**

```python
import pandas_ta as ta
```

```python
stock['EMA_50']=ta.ema(stock.Close, length=50)#sma ema
stock['EMA_20']=ta.ema(stock.Close, length=20)#sma ema
stock['RSI']=ta.rsi(stock.Close, length=14)
my_bbands = ta.bbands(stock.Close, length=20, std=2)
stock=stock.join(my_bbands)
```

```python
temp = ta.macd(stock.Close)
stock=stock.join(temp)
stock.drop(columns = ['MACD_12_26_9','MACDs_12_26_9','BBP_20_2.0','BBB_20_2.0','BBM_20_2.0'],inplace = True)
```

```python
temp = ta.cci(stock.High,stock.Low,stock.Close)
stock=stock.join(temp)
```

```python
temp = ta.stoch(stock.High,stock.Low,stock.Close,k=10)
stock=stock.join(temp)
stock.drop(columns = ['STOCHd_10_3_3'],inplace = True)
```

```python
temp = ta.supertrend(stock.High,stock.Low,stock.Close)
stock=stock.join(temp)
stock.drop(columns = ['SUPERTl_7_3.0','SUPERTs_7_3.0'],inplace = True)
```

```python
temp = ta.donchian(stock.High,stock.Low)
stock=stock.join(temp)
stock.drop(columns = ['DCM_20_20'],inplace = True)
```

```python
temp = ta.adx(stock.High,stock.Low,stock.Close)
stock=stock.join(temp)
stock.drop(columns = ['DMP_14','DMN_14'],inplace = True)
stock
```

**Instead of using absolute value we are going to use Slope of this indicators**

```python
temp = ta.slope(stock.EMA_50)
stock=stock.join(temp)
stock.rename(columns = {'SLOPE_1' : 'EMA_50_slope'},inplace = True)

temp = ta.slope(stock.EMA_20)
stock=stock.join(temp)
stock.rename(columns = {'SLOPE_1' : 'EMA_20_slope'}, inplace = True)

temp = ta.slope(stock['BBL_20_2.0'])
stock = stock.join(temp)
stock.rename(columns = {'SLOPE_1' : 'BBL_20_2.0_slope'}, inplace = True)

temp = ta.slope(stock['BBU_20_2.0'])
stock = stock.join(temp)
stock.rename(columns = {'SLOPE_1' : 'BBU_20_2.0_slope'}, inplace = True)

temp = ta.slope(stock.DCL_20_20)
stock=stock.join(temp)
stock.rename(columns = {'SLOPE_1' : 'DCL_20_20_slope'},inplace = True)

temp = ta.slope(stock.DCU_20_20)
stock=stock.join(temp)
stock.rename(columns = {'SLOPE_1' : 'DCU_20_20_slope'},inplace = True)

stock.drop(columns = ['EMA_50','EMA_20','DCL_20_20','DCU_20_20','BBL_20_2.0','BBU_20_2.0'], inplace = True)
# print(stock)
stock
```

**Now with help of ATR calculating whether we should have bought or sold**

```python
import numpy as np
import warnings
warnings.filterwarnings("ignore", category=pd.errors.SettingWithCopyWarning)  ##ChainedAssignmentError
warnings.filterwarnings("ignore", category=pd.errors.ChainedAssignmentError)

temp = ta.atr(stock.High,stock.Low,stock.Close)
stock=stock.join(temp)
np_array = np.zeros(len(stock))
stock['Action'] = np_array

for i in range(0,len(stock)):
    for j in range(i,len(stock)):
        temp = (stock['Close'].iloc[j] - stock['Close'].iloc[i])
        if temp>= 3*stock['ATRr_14'].iloc[i]:
            # stock['Action'][i] = 1
            stock.loc[i, 'Action'] = 1
            break
        if temp<=-3*stock['ATRr_14'].iloc[i]:
            # stock['Action'][i] = 0
            stock.loc[i, 'Action'] = 0
            break

stock.dropna(inplace = True)
stock
```

```python
stock.drop(columns = ['Open','High','Low','Volume','Close','SUPERT_7_3.0','ATRr_14'],inplace = True)
stock
```

```python
##Now Instead of passing data of just one day we are going to pass it data of some back days also let say 30 backcandles
```

```python
def createNewDataWhichTakeBackCandlesInAccount(data_stock, backcandles = 30):
    X = []
    #print(data_set_scaled[0].size)
    #data_set_scaled=data_set.values
    backcandles = 30
    print(data_stock.shape[0])
    for j in range(12):#data_set_scaled[0].size):#2 columns are target not X
        X.append([])
        for i in range(backcandles, data_stock.shape[0]):#backcandles+2
            X[j].append(data_stock[i-backcandles:i, j])
    #move axis from 0 to position 2
    X=np.moveaxis(X, [0], [2])
    #Erase first elements of y because of backcandles to match X length
    #del(yi[0:backcandles])
    #X, yi = np.array(X), np.array(yi)
    # Choose -1 for last column, classification else -2...
    X, yi =np.array(X), np.array(data_stock[backcandles:,-1])
    y=np.reshape(yi,(len(yi),1))
    return X,y
```

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler(feature_range=(0,1))

sc.fit(stock)
stock = sc.transform(stock)
test_stock = sc.transform(test_stock)

X_train,y_train = createNewDataWhichTakeBackCandlesInAccount(stock)
X_test,y_test = createNewDataWhichTakeBackCandlesInAccount(test_stock)
```

```python
from keras.models import Sequential
from keras.layers import LSTM
from keras.layers import Dropout
from keras.layers import Dense
from keras.layers import TimeDistributed
from keras.activations import sigmoid
from sklearn.metrics import classification_report
import tensorflow as tf
import keras
from keras import optimizers
from keras.callbacks import History
from keras.models import Model
from keras.layers import Dense, Dropout, LSTM, Input, Activation, concatenate
import numpy as np


model = keras.Sequential([
    keras.layers.Flatten(input_shape = (30,12)),
    keras.layers.Dense(5000,activation = 'relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(1000,activation = 'relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(500,activation = 'relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(1,activation = 'sigmoid')
])

model.compile(loss = 'binary_crossentropy',optimizer = 'adam', metrics = ['accuracy'])
model.fit(x = X_train,y = y_train, batch_size=200, epochs=100,validation_split=0.2)
```

```python
y_pred = model.predict(X_test)
#y_pred=np.where(y_pred > 0.43, 1,0)
# y_pred
real_y_pred = []
for element in y_pred:
  if element>.5:
    real_y_pred.append(1)
  else:
    real_y_pred.append(0)

print(classification_report(y_test, np.array(real_y_pred)))
```

## CNN

```python
model = keras.Sequential([
    keras.layers.Conv2D(filters = 12, kernel_size = (3,3), activation = 'relu', input_shape = (30,12,1)),
    keras.layers.MaxPooling2D((2,2)),

    keras.layers.Conv2D(filters = 13, kernel_size = (3,3), activation = 'relu'),
    keras.layers.MaxPooling2D((2,2)),
    
    keras.layers.Flatten(),
    keras.layers.Dense(5000,activation = 'relu'),
    # keras.layers.Dropout(0.5),
    keras.layers.Dense(1000,activation = 'relu'),
    # keras.layers.Dropout(0.5),
    keras.layers.Dense(100,activation = 'relu'),
    # keras.layers.Dropout(0.5),
    keras.layers.Dense(1,activation = 'sigmoid')
])

model.compile(loss = 'binary_crossentropy',optimizer = 'adam', metrics = ['accuracy'])
model.fit(x = X_train,y = y_train, batch_size=240, epochs=10,validation_split=0.2)
```

```python
y_pred = model.predict(X_test)
#y_pred=np.where(y_pred > 0.43, 1,0)
# y_pred
real_y_pred = []
for element in y_pred:
  if element>.5:
    real_y_pred.append(1)
  else:
    real_y_pred.append(0)

print(classification_report(y_test, np.array(real_y_pred)))
```

## LSTM(RNN)

```python
lstm_input = Input(shape=(30, 12), name='lstm_input')
lstm_layer = LSTM(1000, name='first_layer', dropout=0.33, recurrent_dropout=0.3)(lstm_input)
dense_layer = Dense(500, activation='relu', name='dense_layer')(lstm_layer)
dropout_layer = Dropout(0.4)(dense_layer)  # Adding dropout to the Dense layer
dense_layer = Dense(500, activation='relu', name='dense_layr')(dropout_layer)
dropout_layer = Dropout(0.4)(dense_layer)  # Adding dropout to the Dense layer
# Output layer
output = Dense(1, activation='sigmoid')(dropout_layer)

model = Model(inputs=lstm_input, outputs=output)
adam = optimizers.Adam()
model.compile(optimizer=adam, loss='binary_crossentropy', metrics=['accuracy'])
model.fit(x=X_train, y=y_train, batch_size=240, epochs=100, validation_split=0.2)
```

```python
from sklearn.metrics import classification_report
y_pred = model.predict(X_test)
#y_pred=np.where(y_pred > 0.43, 1,0)
real_y_pred = []
for element in y_pred:
  if element>.5:
    real_y_pred.append(1)
  else:
    real_y_pred.append(0)

print(classification_report(y_test, np.array(real_y_pred)))
```
